```python
# Crie um sistema de gerenciamento de estoque
# funcionalidades: Criar alterar e exibir produtios

print("Bem Vindo ao seu sistema de gerenciamento de estoque!!")
print("Comandos disponiveis: 'criar', 'alterar', 'exibir', 'deletar' 'sair' ")

produtos = ["Supertramp","Aerosmith", "Abba", "Pholas", "Scorpions", "Prince", "AC/DC", "Secos e Molhados" ]

def criar():
  nome_produto = input("Qual o nome do novo disco? ")
  produtos.append(nome_produto)
  print("Produto Criado!")

def alterar():
  print("Produto Alterado!")

def exibir():
  print("Produtos")
  print(produtos)

def deletar():
  print("Produto Deletado!")
  print(produtos)

programa_encerrado = False

def sair():
#  programa_encerrado = True

  while not programa_encerrado:
    acao = input("O que deseja fazer? ")

    if acao == "criar":
      criar()
    elif acao =="alterar":
      alterar()
    elif acao == "exibir":
      exibir()
    elif acao =="sair":
      programa_encerrado = True
      print("Programa encerrado. Até mais!!")

    else:
      print("Opção invalida. Comandos disponiveis : 'criar', 'alterar', 'exibir', 'deletar' 'sair' ")


```
